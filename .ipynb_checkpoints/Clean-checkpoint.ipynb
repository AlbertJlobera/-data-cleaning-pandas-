```python


    
def Seasons(Jan,Feb,May,Apr,Mar,Jun,Jul,Aug,Sep,Octt,Nov,Dec):
    
    winter = "Jan","Feb","Dic"
    spring = "May", "Apr","Mar"
    summer = "Jun","Jul","Aug"
    autumn = "Sep","Oct","Nov"
    
    winter = Jan+Feb+Dec
    spring = May+Apr+Mar
    summer = Jun+Jul+Aug
    autumn = Sep+Octt+Nov
    
    total= winter+spring + summer + autumn
    
    winter=round(winter * 100 / total,2)
    spring= round(spring*100/total,2)
    summer=round(summer * 100 / total,2)
    autumn = round(autumn *100 / total,2)
    return f"Total winter {winter}% total spring {spring}% total summer {summer}% total autumn {autumn}%"

Seasons(5,11,54,63,38,58,64,80,93,60,32,8)    
    
```

```python
import matplotlib.pyplot as plt
import pandas as pd
   
Data = {'Season': ["winter","spring","summer","autumn"],
        'Shark_Attacks': [4.77,27.63,32.8,34.79]       
       }
  
cd = pd(Data,columns=['Season','Shark_Attacks'])
cd.plot(x ='Season', y='Shark_Attacks', kind = 'line')
plt.show()
```

```python
Data = {'Seasons': [4.77,27.63,32.8,34.79]}
df = pd(Data,columns=['Seasons'],index = ["winter","spring","summer","autumn"])

df.plot.pie(y='Seasons',figsize=(5, 5),autopct='%1.1f%%', startangle=90)
plt.show()
```
