## Using Regex:

```python
#Creating new column 'Month':
dc['Month']=None


# Using regex to copy the month ex:'Jul':

index_Date=dc.columns.get_loc('Date')

index_Month=dc.columns.get_loc('Month')

print(index_Date, index_Month)
```

```python

date_pattern = r'[A-Z]{1}[a-z]{2}'
```

```python
#function to add regex result to new column 'Month':

for row in range(0,len(dc)):
    Month = re.search(date_pattern,dc.iat[row,index_Date]).group()
    
    dc.iat[row,index_Month]= Month
dc["Month"].value_counts()
```

```python
# check if all characters in each string in series are digits.

dc = dc[dc['Age'].apply(lambda x: str(x).isdigit())]

# Convert all characters to integers.

dc['Age']=pd.to_numeric(dc['Age'], downcast='integer', errors='coerce')
```

```python
# Substitute the value int to str depending of a condition.

def fn(value):
    return "Adult" if value >= 18 else "Teenager"

dc["Age"] = dc["Age"].apply(fn)
dc
```

```python
# changing Type's of the column 'Year' float to int

dc["Year"] = dc.Year.astype(int)

dc
```

```python
# Sorting column 'Age'.
dc = dc.sort_values(by ='Year' , ascending=False)
 
```
